### Definición de funciones

Las funciones se definen con la palabra clave `def`

```python
def funcion(argumento1, argumento2, argumento_con_default=None, 
            *argumentos_posicionales_aninimos, **otros_argumentos_con_nombre)
    # hacer algo
    algo = 0
    return algo
```

````python
def fun():
    print("hago cuentas")
    return 1, 2, 3
    
print("fuera de funcion")
````

````python
fun()
````

````python
def funcionCompleja(a, b=1, *args, **kwargs):
    "Esta funcion multiplica el arg a x el arg b y luego"
    print("a =", a)
    print("b =", b)
    for i, arg in enumerate(args):
        print("elemento {0} de args es igual a {1}".format(i, arg))
    for i, arg in enumerate(kwargs):
        print("elemento {0} de kwargs es igual a {1}".format(arg, kwargs[arg]))
    return None
````

````python
funcionCompleja(10)
````

En Python, ***args** y ****kwargs** son parámetros especiales que se pueden utilizar en la definición de funciones para permitir la pasada de un número variable de argumentos. Estos parámetros se utilizan a menudo cuando no sabemos de antemano cuántos argumentos se van a pasar a la función.

***args** se utiliza para pasar un número variable de argumentos posicionales a una función. ***args** permite que la función acepte cualquier número de argumentos posicionales. Dentro de la función, ***args** se comporta como una tupla de argumentos posicionales, lo que significa que puede ser iterado, desempaquetado, o pasado a otra función como argumento.

````python
# Ejemplo de *args
def suma(*args):
    total = 0
    for num in args:
        total += num
    return total

print(suma(1, 2, 3)) # 6
print(suma(4, 5, 6, 7)) # 22
````

****kwargs** se utiliza para pasar un número variable de argumentos con nombre a una función. En decir, ****kwargs** permite que la función acepte cualquier número de argumentos con nombre. Dentro de la función, ****kwargs** se comporta como un diccionario de argumentos con nombre, lo que significa que puede ser iterado, desempaquetado, o pasado a otra función como argumento.

````python
# Ejemplo de **kwargs
def imprimir_nombres(**kwargs):
    for nombre, edad in kwargs.items():
        print(f"{nombre} tiene {edad} años.")

imprimir_nombres(Juan=30, Maria=40, Pedro=20)
````

### Definición de Clases

Python admite dos paradigmas de programación:
*   Programación Estructurada 
*   Programación Orientada a Objetos 

Para lo segundo, introducimos el concepto de **Clases**


A partir de una clase creamos determinados **Objetos** con caracteristicas definidas. 

Dichos Objetos son entidades que tienen un *Comportamiento*; un *Estado*; *Almacenan Información* y pueden *Realizar Tareas*.


---

Ejemplo simple para los que no vienen de formación en Programación:

Tenemos la Clase: **Animal**

El objeto instanciado (creado a partir de esa clase): **Perro**

Que tiene sus atributos:
*   Color: negro
*   Raza: Coquer
*   Tiene collar: True

Y tiene sus métodos:
*   Ladra (“Guau; guau”)
*   Corre ()
*   Camina()
*   Olfatea()

---



Las clases se definen con la palabra clave `class`, los argumentos de una clase tienen que ser otras clases de las cuales hereda

````python
class perro(object):
    """
    perro: define una clase perro de color Color y edad Edad
    """
    def __init__(self, Nombre, Color, Edad):
        """
        el método __init__ es el que se llama al crear una instancia de la clase, 
        tiene ciertas similitudes con un constructor en otros lenguajes.
        """
        self.Nombre = Nombre
        self.Color = Color
        self.Edad = Edad

    def presentar(self):
        """
        Presentar: imprime los datos de la clase: sus propiedades o atributos
        """
        return f"Soy un perro llamado {self.Nombre}, de color {self.Color} y de {self.Edad} años de edad"
````

````python
class MiClase(object):
    """
    MiClase: soy una clase que hace .... 
    """
    
    def __init__(self, arg1=None, arg2=None):
        """
        el método __init__ es el que se llama al crear una instancia de la clase
        """
        self.arg1 = arg1
        self.arg2 = arg2

    def sumar(self):
        """
        sumar: suma arg1 + arg2
        """
        return self.arg1 + self.arg2
   
````

````python
instancia = MiClase(15, 1)
````

````python
instancia.sumar()
````

#### Generadores

Los generadores son un tipo especial de iteradores. A diferencia de una lista, por ejemplo, no contienen en memoria todos sus elementos, sino que los van "generando" a medida que le son requeridos

````python
generator = (a for a in range(10))
generator
````

````python
for i in generator:
    print(i)
````

````python
list(generator)
````

Usando yield obtenemos un resultado similar a usar "return", excepto que se guardará el estado de la misma para ser usada como un generador

````python
def gen_return():
    for a in range(10):
        return a
````

````python
generator_return = gen_return()
generator_return
````

````python
type(generator_return)
````

````python
def gen():
    for a in range(10):
        yield a
````

````python
generator = gen()
````

````python
type(gen)
````

````python
type(generator)
````

````python
for a in generator:
    print(a)
````

````python
generador = gen()
````

````python
next(generador)
````

````python
from time import time
def gen():
    while True:
        yield time()
````

````python
for i, t in enumerate(gen()):
    print(i, t)
    if i > 10:
        break
````
